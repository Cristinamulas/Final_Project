```python
# Change the world supressed ( deahts btw 0-9) to 0 maybe change by the mean
df = df.replace('Suppressed', 0)
```

```python
# Functions
def subseting (df , x , y):
    df_copy = df.copy()
    df_copy = df_copy.iloc[ x: y , :]
    df_copy = df_copy[[ 'Year' , 'Deaths']]
    return df_copy
```

```python
df_Alabama = subseting(df, 0 , 16)
df_Alabama
df_Alabama.head()
```

```python
import datetime
```

```python
def convert_to_dt(row):
    dt = datetime.datetime(row.Year, 1, 1)
    return dt
```

```python
df_Alabama.Year = df_Alabama.apply(convert_to_dt, axis=1)
# df_Alabama.info()
# df_Alabama = df_Alabama.set_index('Year', inplace=True)
# df_Alabama.head()
df_Alabama
```

```python
df_Alabama.Year
```

```python
df_Alabama.set_index('Year', inplace = True)
```

```python
df_Alabama.set_index('Year', inplace=True)
```

```python
df_Alabama.plot(figsize=(18,6))
plt.xlabel('Year', fontsize=14)
plt.ylabel('Deaths', fontsize = 15)
plt.show()
```

```python
df_Alabama = subseting(df, 0 , 15)
df_Alabama
df_Alabama.Year.index = pd.to_datetime(df_Alabama.Year.index )
# len(df_Alabama.Year)
# df_Alabama.head()
df_Alabama
```

```python
df_Alabama.plot(figsize=(18,6))
plt.xlabel('Year', fontsize=14)
plt.ylabel('Deaths', fontsize = 15)
plt.show()
```

```python

# Set frequency of time series
df_Alabama = df_Alabama.asfreq(freq='1D')

# Sort the values
df_Alabama = df_Alabama.sort_index(ascending = True)
df_Alabama
```
